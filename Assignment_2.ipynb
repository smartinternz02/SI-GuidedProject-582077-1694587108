### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Table

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('car_crashes')
df.head()
```

### Visualization using Matplotlib and Seaborn

```python
df.shape
```

### Univariate

#### Histogram showing ins_premium values

```python
plt.hist(df['ins_premium'])
```

#### Histplot() for ins_losses

```python
sns.distplot(df['ins_losses'])
```

### Bivariate

##### Using scatterplot between total and alcohol 

```python
y=df['total']
x=df['alcohol']
plt.scatter(x,y)
plt.xlabel('Alcohol')
plt.ylabel('Total')
```

```python
#### Lineplot between total and not_distracted
```

```python
sns.lineplot(x='not_distracted',y='total',data=df,ci=None)
```

#### Barplot of total and abbreviation

```python
sns.barplot(data=df,x='abbrev',y='total')
```

#### Jointplot between no_previous and total

```python
sns.jointplot(x='no_previous',y='total',data=df)
```

##### Using scatterplot between total and alcohol, speeding and not distracted

```python
y=df['total']
x1=df['alcohol']
x2=df['speeding']
x3=df['not_distracted']

plt.scatter(x1,y,color='hotpink')
plt.scatter(x2,y,color='green')
plt.scatter(x3,y,color='cyan')
plt.show()
```

### Multivariate Visualization
